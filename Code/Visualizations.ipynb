```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from imblearn.under_sampling import TomekLinks
```

```python
train=pd.read_csv(r"train.csv")
test=pd.read_csv(r"test.csv")
```

```python
y=train.Label.value_counts(normalize=True).values
x=train.Label.value_counts().index
```

```python
import matplotlib.ticker as mtick
range_=range(len(x))
plt.figure(figsize=(4, 3))
plt.bar(range_ , y)
plt.xticks(range_ , x)
plt.title('Distribution of classes')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
plt.savefig("classes.png",bbox_inches='tight');
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.scatterplot(data=train, x="lon", y="lat")
plt.title('Scatter plot of the lat & lon: Training data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig("loc train.png",bbox_inches='tight');
```

```python
plt.figure(figsize=(4, 3))
sns.scatterplot(data=test, x="lon", y="lat")
plt.title('Scatter plot of the lat & lon: Test data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig("loc test.png",bbox_inches='tight');
```

```python
#TRY to improve model selection with feature selection

corr_matrix = train.iloc[:,1:-2].corr()

plt.figure(figsize=(9,7))
plt.title('Correlation matrix')
sns.heatmap(corr_matrix, annot=True)
plt.savefig("Correlation.png",bbox_inches='tight');
```
